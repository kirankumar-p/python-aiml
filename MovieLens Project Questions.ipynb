```python
tempDf1=moviesDf
tempDf1=tempDf1.drop(['movie id','movie title','unknown'],axis=1)
tempDf1['release year']=pd.DatetimeIndex(tempDf1['release date']).year
tempDf1=tempDf1.drop(['release date'],axis=1)
tempDf1.head()
tempDf1=tempDf1.groupby(['release year']).sum()
tempDf1
tempDf1=tempDf1.div(tempDf1.sum(axis = 1, skipna = True) , axis= 0) * 100
fig,ax = plt.subplots(figsize=(20, 20)) 
plt.xticks(rotation=90)
sns.heatmap(tempDf1, ax=ax, annot=True, linewidths=1, fmt= '.2f',cmap="magma") # the color intensity is based on 
plt.show()
#sns.heatmap(tempDf1,  annot=True, linewidths=0.05, fmt= '.2f',cmap="magma")
```

### 9. Find the top 25 movies according to average ratings such that each movie has number of ratings more than 100 - 10 marks

Hints : 

1. Find the count of ratings and average ratings for every movie.
2. Slice the movies which have ratings more than 100.
3. Sort values according to average rating such that movie which highest rating is on top.
4. Select top 25 movies.
5. You will have to use the .merge() function to get the movie titles.

Note: This question will need you to research about groupby and apply your findings. You can find more on groupby on https://realpython.com/pandas-groupby/.

```python
temp2Df=user_rating.drop(['user id','timestamp'],axis=1)
temp3Df=temp2Df.groupby(['movie id'])['rating'].mean().reset_index(name='avg')
temp4Df=temp2Df.groupby(['movie id'])['rating'].count().reset_index(name='cnt')
temp5Df=temp3Df.set_index('movie id').join(temp4Df.set_index('movie id'))
temp5Df=temp5Df.where(temp5Df['cnt']>100)
temp5Df=temp5Df.dropna().sort_values(by=['avg'], ascending=False).head(25)
temp6Df=moviesDf[['movie id','movie title']]
pd.merge(temp5Df,temp6Df,on='movie id')

# Below are the 25 with movies according to average ratings such that each movie has number of ratings more than 100 
```

### 10. See gender distribution across different genres check for the validity of the below statements - 10 marks

* Men watch more drama than women
* Women watch more Sci-Fi than men
* Men watch more Romance than women

**compare the percentages**

1. Merge all the datasets

2. There is no need to conduct statistical tests around this. Just **compare the percentages** and comment on the validity of the above statements.

3. you might want ot use the .sum(), .div() function here.

4. Use number of ratings to validate the numbers. For example, if out of 4000 ratings received by women, 3000 are for drama, we will assume that 75% of the women watch drama.

```python
# Merging 3 data sets in to finalTemo2
finalTemp1=pd.merge(user_rating,users, on='user id')
finalTemp2=pd.merge(finalTemp1,moviesDf, on='movie id')



#Storing sum for each genre after grouping by gender
finalTemp3=finalTemp2.groupby('gender').sum()


# Storing count for each genre after grouping by gender
finalTemp5=finalTemp2.groupby('gender').count()

# storing total unique gender count in total column
finalTemp3['total']=finalTemp5['user id']

#calculating % for each gener
finalTemp4=finalTemp3.div(finalTemp3.total,axis=0)*100

print(finalTemp4[finalTemp4.columns[13:14]])


#We can say first statement Men watch more drama than women is false based on below readings
```
